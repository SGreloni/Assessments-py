```python
name = input("enter your first and last name: ")
first_name = ""
last_name = ""
for char in name:
    if char != " ":
        first_name += char
    else:
        break
for char in name[len(first_name) + 1:]:     #Starts where the previous "for loop" ended and skips de blank space's position
    last_name+= char
    
print(last_name + ", " + first_name)
```

```python
csn = input("enter you comma separated numbers: ")
l = []
n = ""
for char in csn:    
    if char  == ",":
        l.append(int(n))
        n = ""
    elif char == " ":
        continue
    else:
        n+= char
l.append(int(n))
print(l)
```
