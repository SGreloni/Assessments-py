Write a Python Program to implement your own myreduce() function which works exactly like
Python's built-in function reduce()

```python
def myreduce(fun, lis):
    m = lis[0]
    if len(lis) > 1:
        for i in range(1, len(lis)):
            m = fun(m, lis[i])
    return m
```

```python
#Must import functools since reduc() is not any more a bulit-in function
import functools 

lis = [ 1 , 3, 5, 6, 2, ] 
  
# using reduce to compute sum of list 
print ("The sum of the list elements is : ",end="") 
print (functools.reduce(lambda a,b : a+b,lis)) 

# using myreduce to compute sum of list 
print ("The sum of the list elements is : ",end="") 
print (myreduce(lambda a,b : a+b,lis)) 



# using reduce to compute maximum element from list 
print ("The maximum element of the list is : ",end="") 
print (functools.reduce(lambda a,b : a if a > b else b,lis)) 
# using myreduce to compute maximum element from list 
print ("The maximum element of the list is : ",end="") 
print (myreduce(lambda a,b : a if a > b else b,lis)) 
```

Python's built-in function filter()

```python
def myfilter(fun, iterable):
    r = []
    if fun != None:
        boolean_list = []        
        for i in iterable:
            boolean_list.append(fun(i))
    else:
        boolean_list = iterable
        
    for i in range(len(iterable)):
        if boolean_list[i] == True:
            yield (iterable[i])
            
```

Implement a function longestWord() that takes a list of words and returns the longest one

```python
def longestWord(l):
    r =  l[0]
    for word in l:
        if len(l) > len(r):
            r = word
    return r

l = ["Thank", "you", "for", "this", "magnificent", "course"]
longestWord(l)
```

Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.<br>
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5<br>
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.


```python
class triangle:
    def __init__(self):
        pass
    def enter_sides_length(self, side_1 = 0, side_2= 0, side_3= 0):
        side_1 = int(input("Insert the length of side 1: "))
        side_2 = int(input("Insert the length of side 2: "))
        side_3 = int(input("Insert the length of side 3: "))
        self.side_1 = side_1
        self.side_2 = side_2
        self.side_3 = side_3

class area_of_triangle(triangle):
    def __init__(self, *args):
        super(triangle, self).__init__(*args)
    def calculate(self):
        a = self.side_1
        b = self.side_2
        c = self.side_3
        s = (a+b+c)/2
        return (s*(s-a)*(s-b)*(s-c)) ** 0.5

        
    
#Personally, I would have included the calculation of the area as a method (not as a subclass) inside the triangle class but I did it as the task requested. I think doing it this way has no sense.
        
```

Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n.

```python
def filterLongWords(l,n):
    r = []
    for word in l:
        if len(word) > n:
            r.append(word)
    return r


l = ["Thank", "you", "for", "this", "magnificent", "course"]
filterLongWords(l,4)
```

Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words

```python
def lengthOfWords(l):
    m = []
    for word in l:
        m.append(len(word))
    return m

l = ["Thank", "you", "for", "this", "magnificent", "course"]
lengthOfWords(l)
```
