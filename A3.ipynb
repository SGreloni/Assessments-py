Write a function to compute 5/0 and use try/except to catch the exceptions.

```python
def five_over_zero():
    while True:
        try:
            n = int(input("Enter the number for the numerator (it must be 5): "))           
            if n != 5:
                raise ValueError("Numerator must be 5!")
                
            d = int(input("Enter the number for the denominator (it must be 0): "))
            if d != 0:
                raise ValueError("Denominator must be 0!")        
        except Exception as e:    
            print(e)
            continue
        else:
            try:
                n/d
                print("This should never be printed")
                break
            except Exception as e:
                return "We will say this tends to infinite", e  
                break
```

2.<br>
Implement a Python program to generate all sentences where subject is in ["Americans",
"Indians"] and verb is in ["Play", "watch"] and the object is in ["Baseball","cricket"].

```python
subjects=["Americans","Indians"]
verbs=["play","watch"]
objects=["Baseball","Cricket"]

for sub in subjects:
    for verb in verbs:
        for obj in objects:
            print("{} {} {}".format(sub, verb, obj))
```

Write a function so that the columns of the output matrix are powers of the input vector. <br>
The order of the powers is determined by the increasing boolean argument. Specifically, when
increasing is False, the i-th output column is the input vector raised element-wise to the power
of N - i - 1.


```python
def Vandermonde(vector, increasing = True):
    dimensions = [len(vector), len(vector)] #Number of rows and columns being the length of input vector 
    arr = np.ones(dimensions) 
    power = 0 
    for row in arr:
        arr[power] = np.power(row*vector, power)
        power += 1
    
    order = int(increasing)*2-1             #Transform True to 1 and False to -1
    return np.mat(arr[::order].T) 
```

```python
v = [1,2,3,4]
Vandermonde(v, increasing = True)
```
